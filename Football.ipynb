```python
########################################################################
## https://fbref.com
## Premier League Table

html_text = requests.get('https://fbref.com/en/comps/9/Premier-League-Stats').text
soup = BeautifulSoup(html_text, 'lxml')


x = soup.find('table', class_ = "stats_table sortable min_width")


## An array to store table header
headers = []

for i in range(20):
    y = x.find_all('th')[i]
    headers.append(y['aria-label'])  # This shows full text
#     headers.append(y.text)  # This shows abbreviated text
    
    
## Turn headers into dataframe
df = pd.DataFrame(columns = headers)


## Get data from the rest of the table and save it to the previous dataframe
for row in x.find_all('tr')[1:]:
    data = row.find_all(['td', 'th'])
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data


df
```
